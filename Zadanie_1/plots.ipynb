```python
import matplotlib.pyplot as plt

t1 = 9.403924
t2 = 4.459727
t4 = 2.473506
t8 = 1.372586
t12 = 1.032987


procs = [1, 2, 4, 8, 12]
times = [t1, t2, t4, t8, t12]
t1 = times[0]
speedup = [t1 / t for t in times]

plt.figure()
plt.plot(procs, speedup, marker='o')
plt.xlabel("Liczba procesów n")
plt.ylabel("Skalowalność S(n) = czas(1)/czas(n)")
plt.title("Skalowalność programu Monte Carlo pi")
plt.grid(True)
plt.show()
```
